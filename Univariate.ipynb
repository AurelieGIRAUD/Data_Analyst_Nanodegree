## Outliers

<div class="alert alert-block alert-info">
There are many techniques to handle outliers, as our variables seem to follow a Noral distribution, we could for example drop the values above three standard deviations from the median 
(according to the the empirical rule).</div>

```python
def remove_outliers(df):
    outliers = {}
    for i in df.columns:
        df2 = df[np.abs(df[i]-df[i].mean()) < (3*df[i].std())]
        out = df[~(np.abs(df[i]-df[i].mean()) < (3*df[i].std()))]
        out = pd.DataFrame(out)
    return (df2,out)
```
